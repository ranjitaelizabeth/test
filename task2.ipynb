```python
import pandas as pd
import numpy as np
from geopy.distance import distance
import concurrent.futures
```

```python
data = pd.read_csv('corrected_latitude_longitude_details.csv')

```

```python
 def generate_left_right_points(row):
    orig_point = (row.latitude, row.longitude)
    bearing_left = 270  # 270 degrees for left
    bearing_right = 90  # 90 degrees for right

    point_left = distance(meters=25).destination(orig_point, bearing_left)
    point_right = distance(meters=25).destination(orig_point, bearing_right)

    return (row.latitude, row.longitude, point_left.latitude, point_left.longitude, point_right.latitude, point_right.longitude)


```

```python
with concurrent.futures.ThreadPoolExecutor() as executor:
    results = list(executor.map(generate_left_right_points, data.itertuples(index=False)))

# Create a DataFrame for the new points
new_data = pd.DataFrame(results, columns=['orig_latitude', 'orig_longitude', 'left_latitude', 'left_longitude', 'right_latitude', 'right_longitude'])
```

```python
from collections import deque

class MovingTotal:
    def __init__(self):
        self.numbers = deque(maxlen=3)
        self.totals = set()
    
    def append(self, numbers):
        for num in numbers:
            self.numbers.append(num)
            if len(self.numbers) == 3:
                self.totals.add(sum(self.numbers))
    
    def contains(self, total):
        return total in self.totals

if __name__ == "__main__":
    movingtotal = MovingTotal()
    movingtotal.append([1, 2, 3, 4])
    print(movingtotal.contains(6))  # True
    print(movingtotal.contains(9))  # True
    print(movingtotal.contains(12)) # True
    print(movingtotal.contains(7))  # False
    movingtotal.append([5])
    print(movingtotal.contains(6))  # True
    print(movingtotal.contains(9))  # True
    print(movingtotal.contains(12)) # True
    print(movingtotal.contains(7))  # False
```
